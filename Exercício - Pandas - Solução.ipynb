### Qual o gênero de filme mais lucrativo em todo o mundo?

```python
import pandas as pd
import numpy as np
```

```python
movies_df = pd.read_csv("datasets/movies.csv")
movies_df
```

```python
generos = movies_df["Major Genre"].unique()
```

```python
genre_list = [movies_df[movies_df['Major Genre'] == generos[i]]["Worldwide Gross"].sum() for i in range(len(generos))]
```

```python
#movies_df[movies_df["Major Genre"] == "Comedy"]["Worldwide Gross"].sum()
```

```python
genre_list
```

```python
generos
```

```python
results = pd.Series(genre_list, generos, dtype=int)
results
```

```python
results.sort_values(ascending=False)
```

### Quais filmes tiverem os maiores arrecadações (Gross) em relação ao orçamento de produção (Budget)?

```python
movies_df["lucro_mundial"] = movies_df["Worldwide Gross"] - movies_df["Production Budget"]
```

```python
movies_df
```

```python
movies_index = [movies_df["lucro_mundial"].sort_values(ascending=False).index[i] for i in range(10)]
```

```python
movies_df["Title"].loc[movies_index] # Mostrando os 10 primeiros filmes com maiores arrecadações mundialmente
```
